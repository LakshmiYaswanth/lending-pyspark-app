```python
from pyspark.sql import SparkSession  
import getpass 
username = getpass 
spark = SparkSession.builder.config('spark.ui.port','0').\
        config('spark.sql.warehouse.dir',f"/user/itv022063/warehouse").\
        enableHiveSupport().\
        master('yarn').\
        getOrCreate()
```

```python
customerSchema = """
member_id string, emp_title string, emp_length string, 
home_ownership string, annual_inc float, addr_state string, 
zip_code string, country string, grade string, sub_grade string, 
verification_status string, tot_hi_cred_lim float, 
application_type string, annual_inc_joint float, verification_status_joint string
""";
```

```python
customerdf=spark.read \
           .format('csv') \
           .schema(customerSchema)\
           .option("header","true") \
           .load("/user/itv022063/Yash/raw/customer")
```

```python
customerdf.printSchema()
```

```python
customerdf=customerdf.withColumnRenamed("annual_inc","annual_income") \
          .withColumnRenamed("addr_state","address_state") \
          .withColumnRenamed("zip_code","address_zipcode") \
          .withColumnRenamed("country","address_country") \
          .withColumnRenamed("country","address_country") \
          .withColumnRenamed("tot_hi_cred_lim","total_high_credit_limit") \
          .withColumnRenamed("annual_inc_joint","join_annual_income") \
```

```python
from pyspark.sql.functions import *
customerdf=customerdf.withColumn("ingest_date",current_timestamp())
```

```python
display(customerdf)
```

```python
customerdf.filter(col("annual_income").isNotNull())
```

```python
customerdf_emplength=customerdf.withColumn("emp_length", regexp_replace(col("emp_length"), "(\D)", ""))
```

```python
customerdf_emplength=customerdf_emplength.withColumn("emp_length",customerdf_emplength.emp_length.cast("int"))
```

```python
customerdf_emplength.printSchema()
```

```python
avg_emp_length=customerdf_emplength.agg(floor(avg("emp_length"))).collect()[0][0]
display(avg_emp_length)
```

```python
customerdf_emplength=customerdf_emplength.na.fill(avg_emp_length,subset=["emp_length"])
```

```python
customer=customerdf_emplength.withColumn("address_state", when(length(col("address_state")) > 2,
"NA").otherwise(col("address_state")))
```

```python
customer=customer.na.fill("NA",subset=["verification_status_joint"])
```

```python
display(customer)
```

```python
customer.write \
.format("parquet") \
.mode("overwrite") \
.option("path", "/user/itv022063/Yash/raw/customers_parquet") \
.save()
```

```python
customer.write \
.format("csv") \
.mode("overwrite") \
.option("path", "/user/itv022063/Yash/raw/customers_csv") \
.save()
```
